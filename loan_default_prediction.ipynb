```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('Loan_default_kaggle.csv')
df.head()
```

```python
#finding shape of the data
df.shape
```

```python
df.columns
```

```python
# droping unwanted columns
df1 = df.drop(['LoanID', 'Education'], axis = 1)
df1.head()
```

```python
df1.shape
```

```python
# using sklearn label encoder mapping (0,1) HasMortgage, HasDependents, HasCoSigner
encoder = LabelEncoder()
df1['HasMortgage'] = encoder.fit_transform(df1['HasMortgage'])
df1['HasDependents'] = encoder.fit_transform(df1['HasDependents'])
df1['HasCoSigner'] = encoder.fit_transform(df1['HasCoSigner'])
df1['EmploymentType'] = encoder.fit_transform(df1['EmploymentType'])
df1['MaritalStatus'] = encoder.fit_transform(df1['MaritalStatus'])
df1['LoanPurpose'] = encoder.fit_transform(df1['LoanPurpose'])
```

```python
df1.head(15)
```

```python
df.HasCoSigner.isnull().sum()
```

```python
df1.shape
```

```python
X = df1.drop('Default', axis = 1)
X.head()
```

```python
y = df1['Default']
y.head()
```

```python
# train test splitting 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=None, shuffle=False )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
 # importing Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pre = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
